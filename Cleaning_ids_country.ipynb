```python
#Updates the column name by replacing white spaces with "_"
columns=['Country Code', 'Short Name', 'Table Name', 'Long Name', '2-alpha code',
       'Currency Unit', 'Special Notes', 'Region', 'Income Group', 'WB-2 code',
       'National accounts base year', 'National accounts reference year',
       'SNA price valuation', 'Lending category', 'Other groups',
       'System of National Accounts', 'Alternative conversion factor',
       'PPP survey year', 'Balance of Payments Manual in use',
       'External debt Reporting status', 'System of trade',
       'Government Accounting concept', 'IMF data dissemination standard',
       'Latest population census', 'Latest household survey',
       'Source of most recent Income and expenditure data',
       'Vital registration complete', 'Latest agricultural census',
       'Latest industrial data', 'Latest trade data',
       'Latest water withdrawal data']
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
print(df.columns)
```

```python
#Drops unnecessary columns
columns=["Table_Name","Long_Name","2-alpha_code","Special_Notes","WB-2_code","Other_groups","System_of_National_Accounts","Balance_of_Payments_Manual_in_use","Alternative_conversion_factor"]
df.drop(columns,inplace=True,axis=1)
```

```python
#Filling NaN with 0
df.fillna(0)
```

```python
df.replace(r'^\s+$', np.nan, regex=True)
```

```python
#Remove unnecessary values written in brackets
df['SNA_price_valuation'] = df['SNA_price_valuation'].str.replace(r"\(.*\)","")
df['IMF_data_dissemination_standard'] = df['IMF_data_dissemination_standard'].str.replace(r"\(.*\)","")
df['Source_of_most_recent_Income_and_expenditure data'] = df['Source_of_most_recent_Income_and_expenditure_data'].str.replace(r"\(.*\)","")
```

```python
#Splits a column "Source_of_most_recent_Income_and_expenditure_data" into two separate columns for better analysis

df['Source_of_Income'] = df.Source_of_most_recent_Income_and_expenditure_data.str.split(",",expand=True)[0]
df['Expenditure_year'] =df.Source_of_most_recent_Income_and_expenditure_data.str.split(",",expand=True)[1]
```

```python
#Splits a column "Latest_household_survey" into name and year columns
df['Survey_Name'] = df.Latest_household_survey.str.split(",",expand=True)[0]
df['Survey_Year'] = df.Latest_household_survey.str.split(",",expand=True)[1]
```
